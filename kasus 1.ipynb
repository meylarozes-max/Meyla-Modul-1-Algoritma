```python
# Kasus 1 - Kendaraan darat

Jarak = [2000, 1800, 200] # meter
Waktu = [90, 50, 40] # detik
Kecepatan = []

# Hitung kecepatan tiap kendaraan
for i in range(len(Jarak)):
    Kecepatan.append(Jarak[i] / Waktu[i])

# Tampilkan hasil
print("Kecepatan mobil A:", Kecepatan[0], "m/s")
print("Kecepatan motor B:", Kecepatan[1], "m/s")
print("Kecepatan sepeda C:", Kecepatan[2], "m/s")

#Tentukan kendaraan tercepat
tercepat = max(Kecepatan)
print("Kecepatan tertinggi:", tercepat, "m/s")
```
