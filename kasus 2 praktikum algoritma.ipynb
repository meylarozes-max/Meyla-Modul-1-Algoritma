```python
# Kasus 2 - Atlet lari

Jarak = [400, 800, 200] # meter
Waktu = [50,120,11] # detik
Kecepatan = []

# Hitung kecepatan tiap atlet
for i in range(len(Jarak)):
    Kecepatan.append(Jarak[i] / Waktu[i])

# Tampilkan hasil
print("Kecepatan pelari 1:", Kecepatan[0], "m/s")
print("Kecepatan pelari 2:", Kecepatan[1], "m/s")
print("Kecepatan pelari 3:", Kecepatan[2], "m/s")

#Tampilkan pelari tercepat
tercepat = max(Kecepatan)
print("Kecepatan tertinggi:", tercepat, "m/s")
```
