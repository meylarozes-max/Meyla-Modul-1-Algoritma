```python
# Kasus 4 - Udara, Laut, dan Darat

Jarak = [1200, 5000, 60] # Meter
Waktu = [60, 300, 7] # Detik
Kecepatan = []

# Hitung kecepatan tiap kendaraan
for i in range(len(Jarak)):
    Kecepatan.append(Jarak[i] / Waktu[i])

# Tampilkan hasil
print("Kecepatan drone:", Kecepatan[0], "m/s")
print("Kecepatan kapal cepat:", Kecepatan[1], "m/s")
print("Kecepatan pemain sepak bola:", Kecepatan[2], "m/s")

# Tentukan yang tercepat
tercepat = max(Kecepatan)
print("kecepatan tercepat:", tercepat, "m/s")
```
