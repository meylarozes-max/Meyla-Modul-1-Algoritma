```python
# Program Simulasi Gerak Lurus (GLBB)
# Menggunakan Looping, If condition, Function, dan Plot Grafik

import matplotlib.pyplot as plt # Import library matplotlib

# Fungsi perhitungan posisi GLBB
def hitung_posisi(v0, a, t):
  return v0 * t + 0.5 * a * (t ** 2)

print("SIMULASI GERAK LURUS (GLBB)")
v0 = float(input("kecepatan (m/s): "))
a = float(input("percepatan (m/s^2): "))
t_max = int(input("waktu (s): "))

# Looping simulasi berdasarkan waktu
t_values = []
s_values = []

for t in range(t_max + 1): # Looping dari 0 hingga t_max
   posisi = hitung_posisi(v0, a, t)
   t_values.append(t)
   s_values.append(posisi)

   # If condition
   if v0 + a * t < 0:
      print(f"Waktu: {t} s benda berhenti karena kecepatannya 0 m/s.")
      break # menghentikan simulasi jika objek berhenti

# Output hasil simulasi
print("\n===HASIL SIMULASI===")
for i in range(len(t_values)):
    print(f"Waktu: {t_values[i]} s | Posisi: {round(s_values[i], 2)} m")

# Plot Grafik
plt.plot(t_values, s_values, marker='o')
plt.title("Grafik Posisi terhadap Waktu (GLBB)")
plt.xlabel("Waktu (s)")
plt.ylabel("Posisi (m)")
plt.grid(True)
plt.show()

# Kesimpulan sederhana berdasarkan nilai percepatan
if a > 0:
   print("\nBenda mengalami percepatan, sehingga kecepatannya bertambah.")
elif a < 0:
   print("\nBenda mengalami perlambatan, sehingga kecepatannya berkurang.")
else:
   print("\nBenda bergerak dengan kecepatan konstan (Gerak Lurus Beraturan).")
```
