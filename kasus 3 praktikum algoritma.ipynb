```python
# Kasus 3 - Transportasi umum

Jarak = [5000, 3000, 2500] # Meter
Waktu = [180, 120, 120] # Detik
Kecepatan = []

# Hitung kecepatan tiap kendaraan
for i in range(len(Jarak)):
    Kecepatan.append(Jarak[i] / Waktu[i])

# Tampilkan hasil
print("Kecepatan kereta:", Kecepatan[0], "m/s")
print("Kecepatan bus kota:", Kecepatan[1], "m/s")
print("Kecepatan truk:", Kecepatan[2], "m/s")

# Tentukan yang tercepat
tercepat = max(Kecepatan)
print("kecepatan tercepat:", tercepat, "m/s")
```
